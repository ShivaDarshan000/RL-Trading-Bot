```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from google.colab import files

print("Please upload the dataset (e.g., AAPL_2018.csv).")
uploaded = files.upload()
filename = list(uploaded.keys())[0]

print("Loading dataset...")
data = pd.read_csv(filename)

if 'Date' in data.columns:
    data['Date'] = pd.to_datetime(data['Date'])
    data.set_index('Date', inplace=True)

close_prices = data['Close'].values

scaler = MinMaxScaler(feature_range=(0, 1))
scaled_prices = scaler.fit_transform(close_prices.reshape(-1, 1))

window_size = 5
states = []
for i in range(window_size, len(scaled_prices)):
    state = scaled_prices[i - window_size:i, 0]
    states.append(state)

states = np.array(states)
actions = ['Buy', 'Sell', 'Hold']

def calculate_reward(action, current_price, next_price):
    if action == 'Buy':
        return max(0, next_price - current_price)
    elif action == 'Sell':
        return max(0, current_price - next_price)
    else:
        return 0

num_states = states.shape[0]
num_actions = len(actions)
q_table = np.zeros((num_states, num_actions))


learning_rate = 0.1
discount_factor = 0.95
epsilon = 1.0
epsilon_decay = 0.995
min_epsilon = 0.01
episodes = 1000


print("Training the RL agent...")
for episode in range(episodes):
    total_reward = 0
    for state_idx in range(num_states - 1):
        if np.random.rand() < epsilon:
            action_idx = np.random.randint(num_actions)
        else:
            action_idx = np.argmax(q_table[state_idx])

        current_state = states[state_idx]
        next_state_idx = state_idx + 1

        current_price = close_prices[window_size + state_idx]
        next_price = close_prices[window_size + next_state_idx]
        reward = calculate_reward(actions[action_idx], current_price, next_price)
        total_reward += reward

        q_table[state_idx, action_idx] = q_table[state_idx, action_idx] + learning_rate * (
            reward + discount_factor * np.max(q_table[next_state_idx]) - q_table[state_idx, action_idx]
        )


    epsilon = max(min_epsilon, epsilon * epsilon_decay)


    if (episode + 1) % 100 == 0:
        print(f"Episode {episode + 1}/{episodes}, Total Reward: {total_reward}")

print("Training completed!")


print("Simulating trading...")
trading_signals = []
portfolio = 0
for state_idx in range(num_states - 1):
    action_idx = np.argmax(q_table[state_idx])
    action = actions[action_idx]


    if action == 'Buy':
        portfolio += close_prices[window_size + state_idx]
    elif action == 'Sell':
        portfolio -= close_prices[window_size + state_idx]

    trading_signals.append(action)


print(f"Final Portfolio Value: {portfolio}")
print("Generated trading signals:")
print(trading_signals[:10])


plt.figure(figsize=(14, 7))
plt.plot(close_prices[window_size:], label='Actual Close Prices', color='blue')
for i, signal in enumerate(trading_signals):
    if signal == 'Buy':
        plt.scatter(i, close_prices[window_size + i], color='green', label='Buy Signal' if i == 0 else "")
    elif signal == 'Sell':
        plt.scatter(i, close_prices[window_size + i], color='red', label='Sell Signal' if i == 0 else "")
plt.title("Stock Prices with Trading Signals")
plt.xlabel("Time")
plt.ylabel("Price")
plt.legend()
plt.show()
```
